```python
def multiplicar_matrices(m1,m2):
    if len(m1[0])==len(m2):
        m3 = []
        for i in range(len(m1)):
            m3.append([])
            for j in range(len(m2[0])):
                m3[i].append(0)
            for i in range(len(m1)):
                for j in range(len(m2[0])):
                    for k in range(len(m1[0])):
                        #print((i,j),(j,k),(j,k),(k,j))
                        m3[i][j] += m1[i][k]* m2[k][j]
        return m3
    else:
        return None
```
